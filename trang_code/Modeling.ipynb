```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
train = pd.read_csv('../train_model.csv')
train.head()
```

```python
train.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
X = train.drop(columns= ['ID','State.Name','Incidence'])
y = train['Incidence']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
lr.score(X_train,y_train), lr.score(X_test, y_test)
```

```python
test = pd.read_csv('../test_model.csv')
test.drop(columns=['Unnamed: 0'], inplace=True)
test.head()
```

```python
test_predict = test.drop(columns = ['ID','State.Name'])
```

```python
test_preds = lr.predict(test_predict)
```

```python
test_preds_lr = pd.DataFrame({
    'ID': test['ID'],
    'Incidence': test_preds,
})

test_preds_lr.head()
```

```python
test_preds_lr['Incidence'] = test_preds_lr['Incidence'].astype('str')
```

```python
test_preds_lr['ID,Incidence'] = test_preds_lr['ID'] + ',' + test_preds_lr['Incidence']
```

```python
test_preds_lr.set_index('ID')
test_preds_lr
```

```python
test_preds_lr.drop(columns = ['ID','Incidence'])
```

```python
test_preds_lr.to_csv('../lr_model.csv', index=False)
```
