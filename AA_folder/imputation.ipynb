# Imputing with IterativeImputer

```python
train_num_cols = train_df.select_dtypes(include=['number']).columns
train_non_num_cols = train_df.select_dtypes(exclude=['number']).columns
it_imp = IterativeImputer(
    estimator=LinearRegression(),
    max_iter=500
)
it_train = it_imp.fit_transform(train_df[train_num_cols])
it_train = pd.DataFrame(it_train, columns=train_num_cols)
```

```python
train_df_it_imp = pd.concat([it_train, train_df[train_non_num_cols]], axis=1)
```

```python
train_df_it_imp.isnull().sum().sort_values(ascending=False)
```
