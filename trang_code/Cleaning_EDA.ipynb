```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.experimental import enable_iterative_imputer 
from sklearn.impute import IterativeImputer
```

```python
train = pd.read_csv('../train_with_5_parquet.csv')
train.head()
```

```python
train.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
test = pd.read_csv('../test_with_5_parquet.csv')
test.head()
```

```python
test.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
train.isnull().sum()
```

### Frist dataset: fill NA with 0 

```python
train = train.fillna(0)
train.head()
```

```python
test = test.fillna(0)
test.head()
```

```python
train_dummy = pd.get_dummies(train, columns=['Age'], drop_first=True)
train_dummy.head()
```

```python
test_dummy = pd.get_dummies(test, columns=['Age'], drop_first=True)
test_dummy.head()
```

```python
train_dummy.to_csv('../train_model.csv')
```

```python
test_dummy.to_csv('../test_model.csv')
```

### Second dataset: drop columns that have 75% of NA, fill Na with impute

```python
train.dropna(thresh = train.shape[0]*0.75 , axis = 1, inplace = True)
```

```python
train.shape
```

```python
train.info()
```

```python
train_missing = train.drop(columns=['ID', 'State.Name', 'Year', 'Age'], inplace = True)
```

```python
train_missing.reshape(-1,1)
```

```python
it_imp = IterativeImputer(
    estimator=LinearRegression(),
    max_iter=3
)
```

```python
it_inc_missing = it_imp.fit_transform(train_missing)
it_inc_missing = pd.DataFrame(it_inc_missing, columns=train_missing.columns)
```
